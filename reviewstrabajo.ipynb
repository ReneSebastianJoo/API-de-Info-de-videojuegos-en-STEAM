```python
import pandas as pd
import ast
from textblob import TextBlob
from datetime import datetime
```

```python
#se crea el dataframe df_reviews desde su correspondiente archivo json

rows = []

with open ('australian_user_reviews.json', encoding='MacRoman') as f:
    for line in f.readlines():
        rows.append(ast.literal_eval(line))

df_reviews = pd.DataFrame(rows)
```

```python
#tercer paso desanidar reviews


unpacked_data = []
for index, row in df_reviews.iterrows():
    user_id = row['user_id']
    user_url = row['user_url']
    for review_dict in row['reviews']:
        review_dict['user_id'] = user_id
        review_dict['user_url'] = user_url
        unpacked_data.append(review_dict)

df_reviews = pd.DataFrame(unpacked_data)
```

```python
df_reviews
```

```python
def get_sentiment_label(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity
    
    if polarity < -0.2:
        return 0  # Malo
    elif polarity > 0.2:
        return 2  # Positivo
    else:
        return 1  # Neutral

# Aplicar la función a la columna 'review' y crear la nueva columna 'sentiment_analysis'
df_reviews['sentiment_analysis'] = df_reviews['review'].apply(get_sentiment_label)

# Reemplazar la columna 'review' con la nueva columna 'sentiment_analysis'
df_reviews.drop(columns=['review'], inplace=True)

# Mostrar las primeras filas para verificar los cambios
df_reviews.head()
```

```python
#asignando valor 1 a aquellos campos donde la reseña es ausente en columna sentiment_analysis del dataframe reviews

df_reviews['sentiment_analysis'] = df_reviews['sentiment_analysis'].fillna(1)

# Mostrar las primeras filas actualizadas

df_reviews.head()
```

```python
df_reviews = df_reviews.drop('funny', axis=1)
df_reviews = df_reviews.drop('last_edited', axis=1)
```

```python
def transform_date(date_str):
    # Dividir la cadena por espacios
    parts = date_str.split()

    # Obtener el día y el mes
    day = parts[2].replace(',', '')
    month = parts[1]

    # Mapear el nombre del mes a su número correspondiente
    month_dict = {
        'January': '01', 'February': '02', 'March': '03', 'April': '04', 'May': '05', 'June': '06',
        'July': '07', 'August': '08', 'September': '09', 'October': '10', 'November': '11', 'December': '12'
    }
    if month in month_dict:
        month = month_dict[month]

    # Obtener el año si está presente, si no, usar 1900
    year = parts[-1] if len(parts) == 4 else '1900'

    # Formatear la fecha
    formatted_date = f"{year}-{month}-{day}"
    
    return formatted_date

# Aplicar la función a la columna 'posted'
df_reviews['posted'] = df_reviews['posted'].apply(transform_date)
```
